```python
import json

experiment = 'experiment0'

with open('experiments/'+experiment+'/test_'+experiment+'.json') as json_file:
    data = json.load(json_file)

n = 0
T = 0
for example in data:
    if example['exact_match']: T = T+1
    n = n+1

print('\nTest examples:                   {}'.format(n))
print('Test correct examples:           {}'.format(T))
print('Test exact Match:                {}'.format(T/n))


with open('experiments/'+experiment+'/situational_2_'+experiment+'.json') as json_file:
    data = json.load(json_file)
n = 0
T = 0
for example in data:
    if example['exact_match']: T = T+1
    n = n+1
print('\nSituational2 examples:           {}'.format(n))
print('Situational2 correct examples:   {}'.format(T))
print('Situational2 exact Match:        {}'.format(T/n))
```
